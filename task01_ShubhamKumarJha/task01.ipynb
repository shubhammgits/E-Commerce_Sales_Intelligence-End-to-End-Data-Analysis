## Project - 1 : Data Cleaning & Preparation

### Goal:
- Clean a raw dataset by handling missing values, duplicates, and incorrect data.
- Key Requirements:
  - Identify missing or null values
  -  Remove duplicates
  -  Correct data formats (dates,numbers, text)

#### Importing the dataset

```python
import pandas as pd
import numpy as np

df = pd.read_excel('Dataset for Data Analytics.xlsx')
df
```

```python
print('--Nulls--')
print(df.isnull().sum())

print('\n--dtypes--')
print(df.dtypes)

print('\n--Duplicates--')
print('Full row duplicates : ', df.duplicated().sum())
print('OrderID duplicates:', df.duplicated(subset=['OrderID']).sum())
```

### Observations from the dataset : 

- No duplicate OrderIDs
- No full row duplicates
- Dates already clean
- 309 null values in CouponCode (Cleaning work required)
- Both Categorical and Numerical columns looks good

#### Phase 1 : Strategic Imputation (handling the gaps) 

- CouponCode nulls doesn't mean 'missing data', this mean the customers used No coupon. So filling it with 'No Coupon' rather than deleting.

```python
df["CouponCode"] = df["CouponCode"].fillna('No Coupon')

df.isnull().sum()
```

#### Phase 2 : Integrity Audit (checking unique IDs)

```python
duplics = df[df.duplicated(subset=['OrderID'])]
print(f'Duplicated OrderIDs : {len(duplics)}')
```

```python
#if duplicated OrderIDs were there then :

if len(duplics) > 0:
    df = df.drop_duplicates(subset = ['OrderID'])
    print(f'Rows After duplicates removed : {len(duplics)}')
else:
    print('All OrderIDs are unique!')
```

#### Phase 3 : Standerdize (speak one language)

```python
#proper text casing on some cols:

for col in ['Product', 'PaymentMethod', 'OrderStatus', 'ReferralSource']:
    df[col] = df[col].str.strip().str.title()

# df['Product'].unique()


#date to ISO 8601 (YYYY-MM-DD)
df['Date'] = pd.to_datetime(df['Date'].dt.strftime('%Y-%m-%d'))


#round off numericals to 2 decimals:
df['UnitPrice'] = df['UnitPrice'].round(2)
df['TotalPrice'] = df['TotalPrice'].round(2)

print('After Standerdization : ')
df.sample()
```

#### Threshold for Project 2 (final verification gate)

```python
print('--Final Audit--')
print('Total rows : ', len(df))
print('Nulls remaining : ', df.isnull().sum().sum())
print('Duplicate OrderIDs : ', df.duplicated(subset='OrderID').sum())
```

#### Saving the cleaned file :

```python
df.to_excel('Dataset_Cleaned.xlsx', index=False)
print('Saved!')
```
